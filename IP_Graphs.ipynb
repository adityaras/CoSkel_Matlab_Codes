```python
coskel_all_models = pd.read_csv("coskel_all_models.csv")
```

```python
coskel_all_models.head()
```

```python
np.array(coskel_all_models.iloc[0][1:7])
```

```python
models_num = len(coskel_all_models)
```

```python
x = [0,1,2,3,4,5]
plt.figure(figsize=(15,10))
plt.title("F-Scores - Final Epochs of Models")
for i in range(models_num):
  plt.plot(np.array(coskel_all_models.iloc[i][1:7]), label=coskel_all_models.iloc[i][0], marker='o')
  for xs,ys in zip(x,np.array(coskel_all_models.iloc[i][1:7])):
    plt.annotate(str(ys), # this is the text
                 (xs,ys), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center')


plt.xticks(x, ["F0","F1","F2","F3","F4","F5"])
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title("JS - Final Epochs of Models")
for i in range(models_num):
  plt.plot(np.array(coskel_all_models.iloc[i][7]), label=coskel_all_models.iloc[i][0], marker='o')
  # for xs,ys in zip(x,np.array(coskel_all_models.iloc[i][1:7])):
  #   plt.annotate(str(ys), # this is the text
  #                (xs,ys), # these are the coordinates to position the label
  #                textcoords="offset points", # how to position the text
  #                xytext=(0,10), # distance from text to points (x,y)
  #                ha='center')


# plt.xticks(x, ["F0","F1","F2","F3","F4","F5"])
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

```python
models_num
```

```python
tnet_v57_58["F0"]
```

```python
x = [f" {2*i + 1}" for i in range(models_num+1)]
plt.figure(figsize=(15,10))
plt.title("TNET_v57-58 - F Score")
plt.plot(tnet_v57_58["F0"], label="F0", marker='o')
plt.plot(tnet_v57_58["F1"], label="F1", marker='o')
plt.plot(tnet_v57_58["F2"], label="F2", marker='o')
plt.plot(tnet_v57_58["F3"], label="F3", marker='o')
plt.plot(tnet_v57_58["F4"], label="F4", marker='o')
plt.plot(tnet_v57_58["F5"], label="F5", marker='o')

plt.xticks([i for i in range(models_num+1)], x)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.xlabel('Epochs -------> ')
plt.show()
```

```python
x = [f" {2*i + 1}" for i in range(models_num+1)]
plt.figure(figsize=(15,10))
plt.title("TNET_v57-58 - JS")
plt.plot(tnet_v57_58["JS"], label="JS", marker='o')

plt.xticks([i for i in range(models_num+1)], x)
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.xlabel('Epochs -------> ')
plt.show()
```
